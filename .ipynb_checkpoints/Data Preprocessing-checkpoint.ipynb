# Function Definitions

```python
def load_cache():
    '''
    Load the cache containing allensdk dataset
    '''
    # this path determines where downloaded data will be stored
    manifest_path = os.path.join('/local1/ecephys_cache_dir/', "manifest.json")
    return EcephysProjectCache.from_warehouse(manifest=manifest_path)

def get_session_ids(cache, stimulus_type='functional_connectivity'):
    '''
    Get a list of session id's in the dataset with [stimulus_type] stimulus shown
    '''
    sessions = cache.get_session_table()
    return sessions[sessions["session_type"] == stimulus_type].index.values

def load_session_by_id(cache, session_id):
    '''
    Get the session data for the given session id from the given data cache
    '''

    print('Loading session {0} ...'.format(session_id), end='')

    sess = cache.get_session_data(session_id)

    print('done')
    return sess

def load_session_by_idx(cache, idx):
    '''
    Given an index between [0, ..., num_sessions], get a session corresponding to index
    return the session data and the session id
    '''

    session_id = get_session_ids(cache)[idx]
    return load_session_by_id(cache, session_id), session_id

def get_neurons_by_region(session, region):
    '''
    Return a list of neuron (unit) id's according to their anatomical region e.g. "VISp" for primary visual cortex
    '''
    return session.units.loc[session.units.ecephys_structure_acronym==region].index.values

def get_stim_experiment_data(session, stim_name='natural_movie_one_more_repeats', region = 'VISp'):
    '''
    Return a pd dataframe of stimulus presentation information (e.g. frame,  time, etc.) for the stimulus [stimulus_name]
    if presented in the session
    '''
    print('Loading experiment stimulus data...', end='')

    stim_table = session.stimulus_presentations
    stim_table = stim_table.loc[stim_table.stimulus_name==stim_name,  :]

    #stim_table.start_time = np.unwrap(stim_table.start_time)
    t_start = stim_table.start_time.values[0]
    t_end = stim_table.start_time.values[-1]

    print('done')

    return {
        'stim_table' : stim_table,
        't_start' : t_start,
        't_end' : t_end,
        'units' : session.units.index.values, #get_neurons_by_region(session, region),
        'refresh_rate' : np.diff(stim_table.start_time.values)[0]
    }

def get_presentation_times(stim_data):
    '''
     Return the times at which each image is presented as an ndarray.
    '''
    t_start = stim_data['stim_table'].start_time.values[0]
    t_end = stim_data['stim_table'].start_time.values[-1]

    ts = stim_data['stim_table'].start_time.values
    ts = ts[ts < t_end]
    ts = ts[ts > t_start]

    return ts

def get_stimulus_frames(cache, stim_name='natural_movie_one_more_repeats', desired_size=None):
    '''
    Return the frames (pixel data) of the stimulus given by [stim_name]. The
    resulting numpy array has shape (num_frames, num_pixels_x, num_pixels_y). If desired size is specified,
    resulting frames are resized/resampled.
    '''


    if stim_name == 'natural_movie_one_more_repeats':
        frames = cache.get_natural_movie_template(1)
    else:
        assert(False), "Bad Stim Name"

    if desired_size:
        resized_frames = np.zeros((frames.shape[0], desired_size[1], desired_size[0]), dtype=np.float32)
        for i in range(frames.shape[0]):
            img = Image.fromarray(frames[i,:,:])
            resized_image = img.resize(size = desired_size)
            resized_frames[i,:,:] = np.asarray(resized_image)
            resized_frames[i,:,:] = int_intensity_to_float(resized_frames[i,:,:].astype(np.uint8))

        return resized_frames
    else:
        return None

def get_presentation_frame_numbers(stim_data):
    '''
    Return which frames were presented at each time step. Provides indexes to access output
    from get_stimulus_frames. Each frame number is between [0, num_frames].
    '''
    return stim_data['stim_table'].frame.values.astype(np.int)

def get_presentation_frame_times(stim_data):
    '''
    Return the times at which each each frame was presented (start time)
    '''
    return stim_data['stim_table'].start_time.values

def int_intensity_to_float(arr, zero_mean = True):
    '''
    Given an numpy array of uint8 integers representing an image with integer values [0, 255],
    map each intensity to the range [0, 1]. If zero_mean is true, the range is [-.5 , .5].
    '''
    assert(arr.dtype == np.uint8), "Expected array type to be uint8 but was {0}".format(arr.dtype)

    float_arr = np.zeros(arr.shape, dtype =  np.float32)
    float_arr = arr.astype(np.float32) / (255)

    if zero_mean:
        float_arr -= np.mean(float_arr)

    return float_arr

def float_intensity_to_int(arr, zero_mean = True):
    '''
    Inverse operation of int_intensity_to_float above
    '''

    int_arr = arr.astype(np.float32)
    if zero_mean:
        int_arr += .5

    int_arr *= 255

    return int_arr.astype(np.uint8)

    def remove_invalid_spike_times(session, spike_times):
        '''
        Remove spikes that occur between any invalid spike times

        Arguments:
            - session : a Ecephys Session Object
            - spike_times : a (num_timesteps, 1) numpy array of spike times over a given time interval
        '''

        clean_spike_times = spike_times.copy()
        assert(np.all(clean_spike_times == spike_times))

        invalid_table = session.invalid_times
        num_invalid_intervals = len(invalid_table.index)

        invalid_starts = invalid_table.start_time.values
        invalid_stops = invalid_table.stop_time.values
        assert(len(invalid_starts) == len(invalid_stops))

        num_removed_spikes = 0
        for j in range(num_invalid_intervals):
            start = invalid_starts[j]
            stop = invalid_stops[j]
            invalid_spikes_mask = (clean_spike_times >= start) & (clean_spike_times < stop)
            clean_spike_times = clean_spike_times[ ~invalid_spikes_mask]
            num_removed_spikes += np.sum(invalid_spikes_mask)
            assert(len(spike_times) - len(clean_spike_times) == num_removed_spikes)

        return clean_spike_times

    print('Extracting spike data ...', end='')
    spike_times = {}


    for unit in units:
        this_unit_spikes = session.spike_times[unit].copy()
        spike_times[unit] = remove_invalid_spike_times(session, this_unit_spikes)

    print(' done')
    return spike_times

def get_spike_data(units, session):
    '''
    Given a session, return the spike times for units (probes)
    as a dictionary with keys [unit_number] and values [list(spike times for unit_number)].
    '''

    def remove_invalid_spike_times(session, spike_times):
        '''
        Remove spikes that occur between any invalid spike times

        Arguments:
            - session : a Ecephys Session Object
            - spike_times : a (num_timesteps, 1) numpy array of spike times over a given time interval
        '''

        clean_spike_times = spike_times.copy()
        assert(np.all(clean_spike_times == spike_times))

        invalid_table = session.invalid_times
        num_invalid_intervals = len(invalid_table.index)

        invalid_starts = invalid_table.start_time.values
        invalid_stops = invalid_table.stop_time.values
        assert(len(invalid_starts) == len(invalid_stops))

        num_removed_spikes = 0
        for j in range(num_invalid_intervals):
            start = invalid_starts[j]
            stop = invalid_stops[j]
            invalid_spikes_mask = (clean_spike_times >= start) & (clean_spike_times < stop)
            clean_spike_times = clean_spike_times[ ~invalid_spikes_mask]
            num_removed_spikes += np.sum(invalid_spikes_mask)
            assert(len(spike_times) - len(clean_spike_times) == num_removed_spikes)

        return clean_spike_times

    print('Extracting spike data ...', end='')
    spike_times = {}


    for unit in units:
        this_unit_spikes = session.spike_times[unit].copy()
        spike_times[unit] = remove_invalid_spike_times(session, this_unit_spikes)

    print(' done')
    return spike_times

def integrate_spikes_over_window(t_start, tau_int, units, spike_times):
    '''
    Compute the spike rates of each unit/neuron over a given time window.

    :param t_start: (float) the start time of the integration window

    :param tau_int: (float) the length of the integration window

    :param units: (integer list) list of id's of each unit/neuron to integrate over

    :param spike_times (dictionary) Dictionary mapping unit id keys to lists of spike times

    :return: spike_rates (numpy array): The spiking rates of each of units over the time window, with shape (len(units),)
    '''

    spike_rates = np.zeros((len(units),))

    for i, unit in enumerate(units):
        spike_rates[i] = np.sum(
            (spike_times[unit] > t_start)
            &
            (spike_times[unit] <= t_start + tau_int)
        ) / tau_int

    return spike_rates

def compute_spike_rates_after_frame_presentation(frame_times, units, session, tau_int = .030):
    '''
    Given a list of frame presentation times, compute the spike rates of the units/neurons after each
    frame is presented over a window with width tau_int.

    :param frame_times: (float list) A list of frame presentation times with shape (num_frames,). frame_times[i] is the start time of the ith frame

    :param units: (integer list) list of id's of each unit/neuron to integrate over

    :param session: (AllenSDK EcephysSession) The recording session containing experiment data

    :param tau_int: (float) the length of the integration window


    :return: spike_rates (numpy array): The spike rates of each unit/neuron for each given presentation time.
    The shape is (num_frames, num_units).
    '''

    print('Computing spike rates over width = {0} ... '.format(tau_int), end='')
    num_frames, num_units = len(frame_times), len(units)

    spike_rates = np.zeros((num_frames, num_units))

    spike_times = get_spike_data(units, session)

    for j in tqdm(range(num_frames)):
        spike_rates[j, :] = integrate_spikes_over_window(frame_times[j], tau_int, units, spike_times)



    print('done')
    return spike_rates

def z_score(arr, axis = 0):
    '''
        Transform 2d array to zero mean unit variance data along the time axis (2nd dim).

    '''

    means = np.mean(arr, axis = 0, keepdims = True)
    stds = np.std(arr, axis = 0, keepdims = True)

    arr -= means
    arr /= stds

    return arr

def z_score_binned_spikes(binned_spikes, jumps):
    '''
    Z score binned spikes along the time axis to have zero mean and unit variance. Performs process over each
    time interval separated by jumps
    '''
    #TODO: Finish Implementation, Now only implemented for one jump
#     binned_spikes[:, :jumps[0][0]] = z_score(binned_spikes[:, :jumps[0][0]])
#     binned_spikes[:, jumps[0][0]:] = z_score(binned_spikes[:, jumps[0][0]:])

    return z_score(binned_spikes)

def align_frames_spike_rates(frame_numbers, movie, zero_mean_pixel_intensity = True, zero_mean_pixel_vs_time = True):
    '''
    todo
    frame_numbers: (list) A list with length (num_frames) of frame numbers indexing pixel data in movie.

    return: stimulus_frames: (numpy array) A numpy array containing the pixel intensities for each frame presented.
    The shape is (num_frames, num_pixels)
    '''


    # Input data is binned spikes shifted by num_timesteps_delay
    num_frames = len(frame_numbers)
    num_pixels = len(movie[0,:,:].flatten())

    stimulus_frames = np.zeros((num_frames, num_pixels))

    for frame_idx in range(num_frames):
        stimulus_frames[frame_idx, :] = movie[frame_numbers[frame_idx], :, :].flatten()

        if zero_mean_pixel_intensity:
            stimulus_frames[frame_idx, :] -= np.mean(stimulus_frames[frame_idx, :])

    if zero_mean_pixel_vs_time:
        stimulus_frames -= np.mean(stimulus_frames, axis = 0)

    return stimulus_frames

def evaluate_filters(X_test, Y_test, filters):
    '''
    Evaluate a set of linear filters on a test set of X_test
    neuron activities, comparing to Y_test true images. Return the
    average percent error over the set
    '''
    num_samples = X_test.shape[0]
    avg_pct_err = 0

    y_hat_test = X_test @ filters

    assert(y_hat_test.shape == Y_test.shape)

    avg_pct_err  = np.divide(np.linalg.norm(y_hat_test - Y_test, axis = 1), np.linalg.norm(Y_test, axis = 1))

    return np.sum(avg_pct_err) / num_samples
```

# Empirically Determine Best Integration Window Width

```python
# Number of times randomly shuffled, train, and evaluate each integration width 
num_shuffles = 10

# potential widths.  The code below assumes you've already ran and saved spike_rates data files for each value of potential widths
potential_widths = np.logspace(-3, -.5, num = 10)

#initialize Training and testing errors
test_errors = np.zeros((num_shuffles, len(potential_widths)))
train_errors = np.zeros(test_errors.shape)


# count for indicating progress
count = 1


# sweep through each potential width 
for i, tau_int in enumerate(potential_widths):

    # load spike rate data saved on hard drive
    file_name = 'spike_rates_width={0}_ms'.format(np.round(1000 * tau_int)) + '.npy'

    # z score (zero mean unit variance) spike data across time 
    spike_rates_z_scored = z_score(np.load(file_name))
        
    # shuffle train/test data [num_shuffles] times
    for j in range(num_shuffles):
        X_train, X_test, Y_train, Y_test = train_test_split(spike_rates_z_scored, stimulus_frames, test_size=0.1, random_state = j)

        # least squares linear filter 
        filters = np.linalg.pinv(X_train) @ Y_train
        
        # evaluate accuracy on train and test data sets 
        test_errors[j,i] = evaluate_filters(X_train, Y_train, filters)
        train_errors[j,i] = evaluate_filters(X_test, Y_test, filters)
        
        # update progress and iterate
        print('Iteration {0} / {1}'.format(count, num_shuffles * len(potential_widths)))
        count += 1
        
        
#plot results with errorbars giving standard deviation over shuffles, centered around mean 
plt.figure(figsize=(16,8))
plt.errorbar(x = 1000 * potential_widths, y = np.mean(train_errors, axis = 0), yerr = np.std(train_errors, axis = 0), label = "Train")
plt.errorbar(x = 1000 * potential_widths, y = np.mean(test_errors, axis = 0),  yerr = np.std(test_errors, axis = 0), label = "Test")
plt.xlabel(r"Integration Window Length $\tau$ (ms)", fontsize=18)
plt.ylabel('Accuracy', fontsize=18)
plt.legend()
plt.title("Linear Decoder Test/Train Accuracy vs. Integration Bin Width", fontsize=18)

plt.savefig('this_fig.png',bbox_inches='tight')
plt.plot()
plt.show()
```
